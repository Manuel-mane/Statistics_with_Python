# Centrality

## Mean


```python
def get_mean(xs):
    summed = 0
    for x in xs:
        summed += x
    return summed/len(xs)

print(get_mean([3,5,2,6]))
```

### We can use two functions

```python
mean = np.mean(data)
print(mean, data.mean(), np.average(data))
```

**Note:** np.average accepts weight, different from np.mean

## Median

Sort all your data and take out the middle element. That's your median. [1,3,5,7,7] has a median of 5. Here is how we can manually find the median

```python
def get_median(xs):
    mid = len(xs)//2
    if len(xs)%2 == 1:
        return sorted(xs)[mid]
    else:
        return 0.5*np.sum(sorted(xs)[mid-1:mid+1])
print(get_median([7,7,3,1,4,5]))
```

```python
np.median(data)
```

**Note:** Median is not sensitive to outliers

## Mode 

Another outlier insensitive method, it reutns to us the value which is most common. **This works for discrete distributions only.. sort of**. If you have a continuous distribution, you will need to bin the data first. For example, the mode of [1,7,2,5,3,3,8,3,2] is 3, because 3 shows up more than any other number. Here is the manual mode function:

```python
def get_mode(xs):
    values, counts = np.unique(xs, return_counts=True)
    max_count_index = np.argmax(counts)
    return values[max_count_index]

print(get_mode([1,7,2,5,3,3,8,3,2]))
```

```python
mode = st.mode(data)
mode
```

```python
hist, edges = np.histogram(data, bins=100)
edge_centers = 0.5*(edges[1:] + edges[:-1])
mode = edge_centers[hist.argmax()]
print(mode)
```

```python
kde = st.gaussian_kde(data)
xvals = np.linspace(data.min(), data.max(), 1000)
yvals = kde(xvals)
mode = xvals[yvals.argmax()]

plt.hist(data, bins=100, density=True, label='data hist', histtype='step')
plt.plot(xvals, yvals, label='KDE')
plt.axvline(mode, label='Mode')
plt.legend();
```

# Measures of width and balance

* Variance
* Standard deviation
* Skewness
* Kurtosis

## Variance

The variance of a distribution is a measure of how much it spreads out around the mean. A touch more formally, its expected value of the squared deviation from the mean. 

```python
def get_variance(xs):
    mean = np.mean(xs)
    summed = 0
    for x in xs:
        summed += (x - mean)**2
    return summed / (len(xs) - 1)
print(get_variance([1,2,3,4,5]))
```

```python
np.var([1,2,3,4,5], ddof=1)
```

```python
np.var(data, ddof=1)
```

## Standard Deviation

Square root of the variance

```python
std = np.std(data)
std, std**2
```

## Our powers combined

```python
xs = np.linspace(data.min(), data.max(), 100)
ys = st.norm.pdf(xs, loc=mean, scale=std)

plt.hist(data, bins=50, density=True, histtype='step', label='Data')
plt.plot(xs, ys, label='Normal approximation')
plt.legend()
plt.ylabel('Probability');
```

## Skewness

There are some standardised ways of quantifying "moments". The first moment is zero by definition. The second is variance. The third is skewness. 

```python
def get_skewness(xs):
    mean = np.mean(xs)
    var = np.var(xs)
    summed = 0
    for x in xs:
        summed += (x - mean)**3
    return (summed/len(xs))/(var**1.5)

print(get_skewness([1,2,3,4,5]))
```

```python
skewness = st.skew(data)
skewness, get_skewness(data)
```

```python
xs = np.linspace(data.min(), data.max(), 100)
ys = st.norm.pdf(xs, loc=mean, scale=std)
ys2 = st.skewnorm.pdf(xs, skewness, loc=mean, scale=std)

plt.hist(data, bins=50, density=True, histtype='step', label='Data')
plt.plot(xs, ys, label='Normal approximation')
plt.plot(xs, ys2, label='Skewnormal approximation')
plt.legend()
plt.ylabel('Probability');
```

```python
xs = np.linspace(data.min(), data.max(), 100)
ys = st.norm.pdf(xs, loc=mean, scale=std)
ps = st.skewnorm.fit(data)
ys2 = st.skewnorm.pdf(xs, *ps)

plt.hist(data, bins=50, density=True, histtype='step', label='Data')
plt.plot(xs, ys, label='Normal approximation')
plt.plot(xs, ys2, label='Skewnormal approximation')
plt.legend()
plt.ylabel('Probability');
```

## Kurtosis

The next moment is a kurtosis

```python
def get_kurtosis(xs):
    mean = np.mean(xs)
    var = np.var(xs)
    summed = 0
    for x in xs:
        summed += (x - mean)**4
    return (summed/len(xs))/(var**2)

print(get_kurtosis([1,2,3,4,5]))
```

```python
kurtosis = st.kurtosis(data, fisher=False) # fisher --> a way of normalizing the distribution
kurtosis, get_kurtosis(data)
```

# Percentiles

```python
ps  =np.linspace(0, 100,30)
x_p = np.percentile(data, ps)

xs = np.sort(data)
ys = np.linspace(0, 1, len(data))

plt.figure(figsize=(10,8))
plt.plot(xs, ys*100, label='ECDF')
plt.plot(x_p, ps, label="Percentiles", marker=".", ms=10)
plt.legend()
plt.ylabel("Percentile");
```

```python
ps = 100*st.norm.cdf(np.linspace(-3, 3, 30))
ps = np.insert(ps, 0, 0)
ps = np.insert(ps, -1, 100)

x_p = np.percentile(data, ps)

xs = np.sort(data)
ys = np.linspace(0, 1, len(data))

plt.figure(figsize=(10,8))
plt.plot(xs, ys*100, label='ECDF')
plt.plot(x_p, ps, label="Percentiles", marker=".", ms=10)
plt.legend()
plt.ylabel("Percentile");
```
