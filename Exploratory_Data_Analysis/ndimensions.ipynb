# N - Dimension

Exploration of higher dimensional data and the relationships between data attributes, I'll load in the file `Diabetes.csv`.

```python
# libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sb
```

```python
df_original = pd.read_csv('./dataset/ND/Diabetes.csv')
df_original.head()
```

```python
cols = [c for c in df_original.columns if c not in ["Pregnancies", "Outcome"]]
df = df_original.copy()
df[cols] = df[cols].replace({0: np.NaN})
df.head()
```

```python
df.info()
```

```python
df.describe()
```

# Scatter Matrix

```python
pd.plotting.scatter_matrix(df, figsize=(15,10));
```

```python
df2 = df.dropna()
colors = df2["Outcome"].map(lambda x: "#44d9ff" if x else "#f95b4a")
pd.plotting.scatter_matrix(df2, figsize=(15,10), color=colors);
```

# Correlation Plots

```python
df.corr()
```

```python
plt.figure(figsize=(10,8))
sb.heatmap(df.corr());
```

```python
plt.figure(figsize=(10,8))
sb.heatmap(df.corr(), annot=True, cmap='viridis', fmt='0.2f');
```

# Histograms

```python
df2 = pd.read_csv("./dataset/ND/height_weight.csv")
df2.info()
df2.describe()
```

```python
plt.figure(figsize=(10,8))
plt.hist2d(df2["height"], df2["weight"], bins=20, cmap='magma')
plt.xlabel("Height")
plt.ylabel("Weight");
```

# Countour Plots

```python
hist, x_edge, y_edge = np.histogram2d(df2["height"], df2["weight"], bins=20)
x_center = 0.5*(x_edge[1:] + x_edge[:-1])
y_center = 0.5*(y_edge[1:] + y_edge[:-1])

plt.figure(figsize=(10,8))
plt.contour(x_center, y_center, hist, levels=4)
plt.xlabel("Height")
plt.ylabel("Weight");
```

# KDE Plots

```python
plt.figure(figsize=(10,8))

sb.kdeplot(df2["height"], df2["weight"], cmap="viridis", bw=0.1)
plt.hist2d(df2["height"], df2["weight"], bins=20, cmap='magma', alpha=0.3);
```

```python
#Smoothing the plot
plt.figure(figsize=(10,8))

sb.kdeplot(df2["height"], df2["weight"], cmap="viridis", bw=2)
plt.hist2d(df2["height"], df2["weight"], bins=20, cmap='magma', alpha=0.3);
```

```python
# Better for presentation
plt.figure(figsize=(10,8))

sb.kdeplot(df2["height"], df2["weight"], cmap="viridis", shade=True)
```

# In Defense of Simplicity

```python
m = df2["sex"] == 1
plt.figure(figsize=(10,8))
plt.scatter(df2.loc[m,"height"], df2.loc[m,"weight"], c="#16c6f7", s=5, label="Male")
plt.scatter(df2.loc[~m,"height"], df2.loc[~m,"weight"], c="#ff8b87", s=5, label="Female")
plt.xlabel("Height")
plt.ylabel("Weight")
plt.legend();
```
